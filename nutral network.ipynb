```
# Defining two fuzzy sets A and B based on Traingular membership function.
import numpy as np
import matplotlib.pyplot as plt

# Define fuzzy sets using membership functions
x = np.arange(0, 10, 0.1)
A = np.maximum(0, np.minimum(1, (x-2)/(6-2)))
B = np.maximum(0, np.minimum(1, (8-x)/(8-4)))

# Plot fuzzy sets A and B
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='Set A')
plt.plot(x, B, label='Set B')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree(lokesh)')
plt.title('Fuzzy Sets A and B')
plt.legend()
plt.show()
```

```
# Complement of fuzzy set A
complement_A = 1 - A

# Plot the result
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='Set A')
plt.plot(x, complement_A, label='Complement of A', linestyle='dashed')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title('Complement of Fuzzy Set A')
plt.legend()
plt.show()
```

```
# General aggregation operation (e.g., Algebraic Sum)
aggregation = A + B - A * B

# Plot the result
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='Set A')
plt.plot(x, B, label='Set B')
plt.plot(x, aggregation, label='Aggregation (Algebraic Sum)', linestyle='dashed')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title('Aggregation Operation on Fuzzy Sets A and B')
plt.legend()
plt.show()
```

```
# Union of fuzzy sets A and B
union = np.maximum(A, B)
# Plot the result
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='Set A')
plt.plot(x, B, label='Set B')
plt.plot(x, union, label='Union A  B', linestyle='dashed')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title('Union of Fuzzy Sets A and B')
plt.legend()
plt.show()
```
